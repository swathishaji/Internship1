```python
import numpy as np

# Clean the train DataFrame
train.replace('?', np.nan, inplace=True)

# Drop rows with missing or invalid values in key columns for train
cols_to_check = ['Health Indicator', 'Holding_Policy_Duration', 'Holding_Policy_Type'] # Assuming these are the key columns based on previous missing value analysis

# Specify invalid values if any (based on previous unique value analysis)
# No obvious invalid values observed in previous unique value analysis for these columns

# Build mask for rows to drop for train
drop_mask_train = train[cols_to_check].isna().any(axis=1)

# Keep only valid rows for train
train = train[~drop_mask_train].reset_index(drop=True)

# Clean the test DataFrame
test.replace('?', np.nan, inplace=True)

# Drop rows with missing or invalid values in key columns for test
# Using the same columns as for train
cols_to_check_test = ['Health Indicator', 'Holding_Policy_Duration', 'Holding_Policy_Type']

# Build mask for rows to drop for test
drop_mask_test = test[cols_to_check_test].isna().any(axis=1)

# Keep only valid rows for test
test = test[~drop_mask_test].reset_index(drop=True)
```

```python
print("Shape of train after cleaning:", train.shape)
print("Shape of test after cleaning:", test.shape)
```

```python
print("Missing values in train dataset (including '?'):")
for col in train.columns:
    q_count = (train[col] == '?').sum()
    nan_count = train[col].isna().sum()
    total_missing = q_count + nan_count
    if total_missing > 0:
        print(f"Column '{col}': '?' = {q_count}, NaN = {nan_count}, Total missing = {total_missing}")

print("\nMissing values in test dataset (including '?'):")
for col in test.columns:
    q_count = (test[col] == '?').sum()
    nan_count = test[col].isna().sum()
    total_missing = q_count + nan_count
    if total_missing > 0:
        print(f"Column '{col}': '?' = {q_count}, NaN = {nan_count}, Total missing = {total_missing}")
```

```python
import numpy as np

print("Missing value proportions in train dataset (including '?'):")
train_missing = train.replace("?", np.nan)
missing_proportions_train = train_missing.isna().mean()

for col, prop in missing_proportions_train.items():
    if prop > 0:
        print(f"Column '{col}': {prop:.2%} missing (including '?')")

print("\nMissing value proportions in test dataset (including '?'):")
test_missing = test.replace("?", np.nan)
missing_proportions_test = test_missing.isna().mean()

for col, prop in missing_proportions_test.items():
    if prop > 0:
        print(f"Column '{col}': {prop:.2%} missing (including '?')")
```

```python
# Replace values in 'Response' column
# Based on previous value counts, 'Response' is already 0 and 1, but applying replacements as requested
train['Response'] = train['Response'].replace(0, 0)
train['Response'] = train['Response'].replace(1, 1)

# Get value counts after replacement
train['Response'].value_counts()
```
