```python
import pandas as pd
df = pd.read_csv('/content/nyc_weather.csv')
df.head()
```

```python
max_temperature = df['Temperature'].max()
max_temperature
```

```python
rainy_days = df[df['Events'] == 'Rain']
rainy_days
```

```python
rainy_days = df[df['Events'].str.contains('Rain',na= False)]
rainy_dates = rainy_days['EST']
display(rainy_dates)
```

```python
average_wind_speed = df['WindSpeedMPH'].mean()
```

```python
print(f"The maximum temperature in new york in january was: {max_temperature}")
print(f"The rainy days in new york in january were: {rainy_dates}")
print(f"The average wind speed in new york in january was: {average_wind_speed}")
```
